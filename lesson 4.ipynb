Случайная непрерывная величина A имеет равномерное распределение на промежутке (200, 800].
Найдите ее среднее значение и дисперсию.

```python
avg_A = (800 + 201) / 2
disp_A = ((800 - 201) ** 2) / 12

print(avg_A, 'среднее значение')
print(disp_A, 'дисперсия')
```

О случайной непрерывной равномерно распределенной величине B известно, что ее дисперсия равна 0.2.
Можно ли найти правую границу величины B и ее среднее значение зная, что левая граница равна 0.5?
Если да, найдите ее.

```python
from math import sqrt

disp_A_2 = 0.2
l_val = 0.5

# disp_A_2 = (r_val - l_val) ** 2 / 12
# 12 * disp_A_2 = (r_val - l_val) ** 2
# sqrt(12 * disp_A_2) = r_val - l_val
# sqrt(12 * disp_A_2) + l_val = r_val

r_val = sqrt(12 * disp_A_2) + l_val
print(r_val, 'правая граница равномерного распределения')
```
